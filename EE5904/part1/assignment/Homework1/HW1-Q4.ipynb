```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

### Q4 a)
LLS algorithm

```python
X = np.array([[1,0],[1,0.8],[1,1.6],[1,3],[1,4],[1,5]])
d = np.array([0.5, 1, 4, 5, 6, 8])

X_t = np.transpose(X)
#Inverse of dot product of XtX
X_t_inv = inv(np.dot(X_t,X))
temp = np.dot(X_t_inv,X_t)
# w = (XtX)-1 Xt d
w = np.dot(temp,d)

#Plot the result
x = np.arange(6)
fig = plt.figure(figsize=(8,4))
graph = fig.add_axes([0,0,1,1])
graph.set_xlabel('Input')
graph.set_ylabel('Output')
graph.plot(x, w[0]+w[1]*x,color='green', label='LLS fitting line')
graph.scatter(X[:,1],d,label='Input data point')
graph.legend(loc='upper left')

print('The final weight vector is:')
print(w)
```

### Q4 b)
LMS algorithm

```python
#Learning rate
lr = 0.01

def LMS(lr):
    X = np.array([[1,0],[1,0.8],[1,1.6],[1,3],[1,4],[1,5]])
    d = np.array([0.5, 1, 4, 5, 6, 8])
    #Initial weight and error
    w = np.random.rand(2)
    rows = np.shape(X)[0]
    e = 0

    #Array to store 100 epochs weight
    weight_array = np.empty([2,100])

    for j in range(100):
        for i in range(rows):
            e = d[i]-np.dot(X[i,:],w)
            w = w + lr*e*X[i,:]
            i = i+1
        weight_array[0,j]=w[0]
        weight_array[1,j]=w[1]
        j=j+1

    print('The final weight vector is:')
    print(w)
    
    #Plot the weight change along 100epochs and LMS fitting line
    step = np.arange(100)
    x = np.arange(6)
    fig, axes = plt.subplots(2, 1, figsize=(8, 8))
    axes[0].plot(step,weight_array[1,:],label='w')
    axes[0].plot(step,weight_array[0,:],label='b')
    axes[0].set_title('LMS weight change along 100 epochs')
    axes[0].set_xlabel('epochs')
    axes[0].set_ylabel('weight')
    axes[0].legend(loc='upper left')
    axes[1].plot(x, w[0]+w[1]*x,label = 'fitting line')
    axes[1].scatter(X[:,1],d,color='green',label = 'input data point')
    axes[1].set_title('LMS fitting line')
    axes[1].set_xlabel('input x')
    axes[1].set_ylabel('desired output d')
    axes[1].legend(loc='upper left')
    plt.tight_layout()
```

```python
LMS(lr=0.01)
```

```python
LMS(lr=0.001)
```
